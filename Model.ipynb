```python
data_train, data_test, target_train, target_test = train_test_split(data,target,test_size=0.2,random_state=10)
```

```python
knn_model = KNeighborsClassifier(n_neighbors=3,metric='manhattan')
knn_model.fit(data_train,target_train)
prediction = knn_model.predict(data_test)
```

```python
accuracy = accuracy_score(target_test, prediction)
```

```python
scores = cross_validate(knn_model,data,target,scoring=['accuracy'])
scores
```

```python
score_result = scores['test_accuracy'].mean() * 100
score_result
```

```python
k_range = list(range(1, 10))
leaf_size = list(range(1, 50))
weight_options = ["uniform", "distance"]
p = [1, 2]
```

```python
param_grid = dict(leaf_size=leaf_size, n_neighbors=k_range, weights=weight_options, p=p)
```

```python
knn = KNeighborsClassifier()
```

```python
grid = GridSearchCV(knn, param_grid, scoring='accuracy')
grid.fit(data, target)

best_score = grid.best_score_
best_params = grid.best_params_
best_estimator = grid.best_estimator_

print('[+] Best Score - ', best_score)
print('[+] Best Params - ', best_params)
print('[+] Best Estimator - ', best_estimator)
```
